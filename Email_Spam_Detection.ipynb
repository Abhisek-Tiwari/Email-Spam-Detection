```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('spam.csv')
df.head()
```

```python
df.Category.value_counts()
```

```python
# convert spam and ham into numbers
df['spam'] = df.Category.apply(lambda x: 1 if x == 'spam' else 0)

df.head()
```

```python
np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.25)
```

```python
# import count vectorizer to make bag of words
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()

X_train_cv = v.fit_transform(X_train.values)
X_train_cv
```

```python
X_train_cv.toarray()
```

```python
X_train_cv.shape
```

```python
v.get_feature_names_out()[3000:3010]
```

```python
# Now we train the Naive Bayes
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(X_train_cv, y_train)
```

```python
X_test_cv = v.transform(X_test)
```

```python
# now time fro evaluation
from sklearn.metrics import classification_report

y_pred = nb.predict(X_test_cv)

print(classification_report(y_test, y_pred))
```

```python
# Lets make a pipeline for automation
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('vectorzer', CountVectorizer()),
    ('nb', MultinomialNB())]
)
```

```python
clf.fit(X_train, y_train)
```

```python
y_preds = clf.predict(X_test)

print(classification_report(y_test, y_preds))
```
